```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_normal = stats.norm.rvs(size=1000, loc=0, scale=1)
print(data_normal)
```

```python
sns.displot(data_normal, bins=100, kde=True)
```

```python
data_uniform = stats.uniform.rvs(size=1000, loc=0, scale=10)
```

```python
sns.distplot(data_uniform, bins=100, kde=False)
```

```python
data_bernouli = stats.bernoulli.rvs(size=1000, p=0.2)
print(data_bernouli)
```

```python
sns.distplot(data_bernouli, bins=100, kde=False)
```

```python
data_binom = stats.binom.rvs(size=1000, n=100, p=0.2)
print(data_binom)
```

```python
sns.distplot(data_binom, bins=50, kde=False)
```

```python
import numpy as np
```

```python
p=1/6
data_multinomial = np.random.multinomial(n=6, pvals=[p, p, p, p, p, p], size=1000)
print(data_multinomial)
```

```python
icu = 0.3
surgery = 0.2
el = 0.5
data_multinomial = np.random.multinomial(n=1000, pvals=[icu, surgery, el], size=1000)
print(data_multinomial)
```

```python
data_poison = np.random.poisson(lam=7, size=1000)
print(data_poison)
```

```python
sns.displot(data_poison)
```

```python
data_exponential = np.random.exponential(scale = 1/2, size=1000)
print(data_exponential)
```

```python
sns.displot(data_exponential)
```
